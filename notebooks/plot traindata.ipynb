## 2D KPE Model

```python
import matplotlib.pyplot as plt
```

```python
# Plot train acc
file_name = "../results/simple_2d_model_20220321-195443_train_acc.txt"
accs = []
with open(file_name) as file:
    accs = file.readlines()
    accs = [float(item.rstrip()) for item in accs]
epochs = [i+1 for i in range(len(accs))]
plt.figure(figsize=(7,6))
plt.plot(epochs, accs)
plt.title("Training 2D KPE Train Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.show()
```

```python
# plot train loss
file_name = "../results/simple_2d_model_20220321-195443_train_loss.txt"
losses= []
with open(file_name) as file:
    losses = file.readlines()
    losses = [float(item.rstrip()) for item in losses]
epochs = [i+1 for i in range(len(accs))]
plt.figure(figsize=(7,6))
plt.plot(epochs, losses)
plt.title("Training 2D KPE Train Losses")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

```python
# plot val acc
file_name = "../results/simple_2d_model_20220321-195443_val_acc.txt"
accs = []
with open(file_name) as file:
    accs = file.readlines()
    accs = [float(item.rstrip()) for item in accs]
epochs = [i+1 for i in range(len(accs))]
plt.figure(figsize=(7,6))
plt.plot(epochs, accs)
plt.title("Training 2D KPE Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.show()
```

```python
# plot val loss
file_name = "../results/simple_2d_model_20220321-195443_val_loss.txt"
losses= []
with open(file_name) as file:
    losses = file.readlines()
    losses = [float(item.rstrip()) for item in losses]
epochs = [i+1 for i in range(len(accs))]
plt.figure(figsize=(7,6))
plt.plot(epochs, losses)
plt.title("Training 2D KPE Validation Losses")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```
